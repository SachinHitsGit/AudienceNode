```python
import sqlite3
from datetime import datetime
import pandas as pd
```

```python
df = pd.read_csv("Amazon_Reviews.csv/Amazon_Reviews.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
req_cols = ["Review Date", "Reviewer Name", "Review Text"]
```

```python
df_to_save = df[req_cols].copy()
df_to_save.columns = ["timestamp", "username", "message"]
df_to_save.index = df_to_save.index + 1
df_to_save.index.name = "id"

db_connection = sqlite3.connect("test_reviews.db")
df_to_save.to_sql("review_logs", db_connection, if_exists="append", index=True)

db_connection.close()
print("Records saved successfully!")
```
